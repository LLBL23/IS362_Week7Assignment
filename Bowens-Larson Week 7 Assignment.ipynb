## Overview
Six recent popular movies were selected and rated on a scale from 1 to 5 by a group of 5 people. The results of the ratings were analyzed to find the average ratings. Then the results were normalized in order to understand their impact on the results.


## Load Data
The information was collected and a .CSV file titled ratings was created containing the source data. The ratings.CSV file was then read into pandas and a DataFrame was created. 

```python
# import pandas
import pandas as pd

# load data from CSV file into a DataFrame
ratings = pd.read_csv("movie_ratings.csv")

# print DataFrame
ratings
```

## Calculate Average Ratings

The data was then seperated into two groups, user and movie, and then the averages were then calculated.

```python
# group by person/user 
user = ratings.groupby('Person')


# group by movie title
movie = ratings.groupby('Movie Title')
```

```python
# print contents of group by looping over all the groups and printing the group name and group details
def print_groups(groupobject):
    for name, group in groupobject:
        print(name)
        print(group)
        
print_groups(user)
print_groups(movie)
```

```python
# calculate the average rating for each user
df_user = user.mean()

# rename rating column
df_user = df_user.rename(columns={'Rating': 'Average_Rating'})

# print DataFrame
df_user
```

```python
# calculate the average rating for each movie
df_movie = movie.mean()

# rename rating column
df_movie = df_movie.rename(columns={'Rating': 'Average_Rating'})

# print DataFrame
df_movie
```

## Normalize Data

A new DataFrame was created containing the normalized ratings for each user. The average ratings were again calulated for each user and each movie. 

```python
# create a function to normalize ratings and create a new column called 'Normalized_Rating' for the data
def normalize_ratings(group):
    min_val = group['Rating'].min()
    max_val = group['Rating'].max()
    # Apply min-max normalization: (Rating - min) / (max - min)
    group['Normalized_Rating'] = (group['Rating'] - min_val) / (max_val - min_val)
    return group
    
# normalize  ratings
ratings_normalized = ratings.groupby('Person').apply(normalize_ratings)

# print DataFrame
ratings_normalized
```

## Calculate Average Normalized Ratings

The normalized data was seperated into two groups, user and movie, and then the averages were then calculated.

```python
# group by person/user
user_normed = ratings_normalized.groupby('Person')


# group by movie title
movie_normed = ratings_normalized.groupby('Movie Title')
```

```python
print_groups(user_normed)
print_groups(movie_normed)
```

```python
# calculate average for each user
df_user_normed_mean = user_normed.mean()

# rename rating columns
df_user_normed_mean= df_user_normed_mean.rename(columns={'Rating': 'Avg_Rating', 'Normalized_Rating': 'Avg_Normalized_Rating'})

# print DataFrame
df_user_normed_mean
```

```python
# calculate average for each movie
df_movie_normed_mean = movie_normed.mean()

# rename rating columns
df_movie_normed_mean= df_movie_normed_mean.rename(columns={'Rating': 'Avg_Rating', 'Normalized_Rating': 'Avg_Normalized_Rating'})

# print DataFrame
df_movie_normed_mean
```

## Conclusion

Advantages of using normalized data is that it can remove outliers, reduce redundancy, and reduce bias by bringing the data to a common scale.

Disadvantages to normalizing data are losing precision and context which can have the result of hiding valuable information in the data.
